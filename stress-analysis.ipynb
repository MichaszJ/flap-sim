```python
import math
import numpy as np
import matplotlib.pyplot as plt
```

```python
# assuming AoA of 2, freestream V of 100, values taken from Project 3 Introduction doc
AoA = 2
CL = 7.6973e-1
CD = 2.5982e-1
CM = -6.9124
V = 100
rho = 1.225
S = 1.3716 * 1.01428

def lift(CL, rho, V, S):
    return CL * 0.5 * rho * math.pow(V, 2) * S

def drag(CD, rho, V, S):
    return CD * 0.5 * rho * math.pow(V, 2) * S
```

```python
# lift and drag used for stress analysis
print('Total lift and drag: ', lift(CL, rho, V, S), drag(CD, rho, V, S))

print('Forces on each piston: ', lift(CL, rho, V, S)/3, drag(CD, rho, V, S)/3)

print('Forces on each rotary actuator: ', lift(CL, rho, V, S)/6, drag(CD, rho, V, S)/6)
```

```python
V_list = np.linspace(10, 300, 500)
L_list = [lift(CL, rho, Vi, S) for Vi in V_list]
D_list = [drag(CD, rho, Vi, S) for Vi in V_list]

plt.style.use('classic')
plt.figure(dpi=230)

plt.plot(V_list, L_list, label='Lift')
plt.plot(V_list, D_list, label='Drag')
plt.legend(loc=2)
plt.xlabel('V (m/s)')
plt.ylabel('L (N)')
plt.show()
```
